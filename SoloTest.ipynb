The init_params functions does the following:
- Create a parameter dictionary that contains weights and baises lists
- Define a lower and upper layer and enumerate through the number of neuron layers 
- Split the key and append onto the weights dictionary jax.normal values given the subkey in the order [higher_layer, lower_layer]
- - Don't forget to divide by the square root of the lower layer for normalization purposes
- Enumerate through the number of neuron layers[1:] and append onto biases jnp.zeroes in the shape num_nerons

The higher layers correspons to the earlier layers of the neural net, so they start from the 2nd onwards given that we predescribe the first layer input and the final layer goes upto the final given that we predescribe the putput layer into the dimention that we're getting to. 

```python
def NN_init_params(key, num_neuron_layers):
    params = {}
    params['w'] = []
    params['b'] = []
    for (higher_layer, lower_layer) in zip(num_neuron_layers[1:], num_neuron_layers[:-1]):
        key, subkey = jax.random.split(key)
        params['w'].append(jax.random.normal(subkey, (higher_layer, lower_layer)) / jnp.sqrt(lower_layer))
    for num_neurons in num_neuron_layers[1:]:
        params['b'].append(jnp.zeros(num_neurons))
    return params
```

Then we define a runge kutta solver and compile it using jax. The way a runge kutta solver works is simple. Consider the quantitiy that you want to numerically integrate. In most cases this is going to be the right hand side of some numerical equation. We then derrive 4 quantities $k_1$, $k_2$, $k_3$, and $k_4$ which are:


\begin{equation}
\begin{aligned}
k_1 &= f(y_n, t_n)\\[6pt]
k_2 &= f\!\left(y_n + \tfrac{h}{2} k_1,\, t_n + \tfrac{h}{2}\right)\\[6pt]
k_3 &= f\!\left(y_n + \tfrac{h}{2} k_2,\, t_n + \tfrac{h}{2}\right)\\[6pt]
k_4 &= f\!\left(y_n + h k_3,\, t_n + h\right)
\tag{1}
\end{aligned}
\end{equation}

Then you step over to the next time step.
\begin{equation}
\begin{aligned}
y_{n+1} &= y_n + \tfrac{h}{6}\bigl(k_1 + 2k_2 + 2k_3 + k_4\bigr)\\[6pt]
t_{n+1} &= t_n + h
\tag{2}
\end{aligned}
\end{equation}

If you were to assume some function *rhs* that takes in an an input a state, some time, and some parameter list then you can define the function as shown in the top half of the code block. 

We want to Jit this in jax though. Given how jax works we need to be able to communicate with it that the *rhs* input is a function as opposed to an array. The way that this is done is by using a @partial command. 

```python
def runge_kutta_step_with_params(X, params, t, rhs, dt):
    """
    Single step definition of the 4th order Runge-Kutta method
    X : State vector
    params : Parameter list
    t : Time
    rhs : Function defining right hand side of ODE
    dt: Timestep

    Important item of note, there are a couple conventions with regards to RK solvers
    You can either return the next step (i.e. X + RK_step)
    or you can return the increment (i.e. RK_step)
    This function definition returns the increment
    """

    k1 = rhs(X, t, params)
    k2 = rhs(X + dt/2 * k1, t + dt/2, params)
    k3 = rhs(X + dt/2 * k2, t + dt/2, params)
    k4 = rhs(X + dt * k3, t + dt, params)
    inc = dt/6 * (k1 + 2*k2+ 2*k3 + k4)
    return inc


@partial(jax.jit, static_argnames=['rhs'])
def runge_kutta_solve(X0, rhs, ts, params):
    """
    Loop definition of the 4th order Runge-Kutta method using the single step definition
    X0 : Initial condition vector
    rhs : Function defining right hand side of ODE
    ts : Time vector
    params : Parameter list

    """
    dt=ts[1]-ts[0]

    def loop_body(x,t):
        dx = runge_kutta_step_with_params(x,params,t,rhs,dt)
        x = x + dx
        return x,x
    
    _,Xs=jax.lax.scan(loop_body,X0,ts[:-1])
    return Xs,ts[1:]

# Batch the above function
batched_param_runge_kutta_solve = jax.vmap(runge_kutta_solve, in_axes=[None, None, None, 0], out_axes=0)

batched_runge_kutta_solve = jax.vmap(runge_kutta_solve, in_axes=[0, None, None, None], out_axes=0)
```

Now we need to pick some function to solve for and define some *rhs* function that takes in as an input a parameter list, state vector, and and time vector. Afterwards we use that rhs function to define another function solve_*equation* that solves the equation using all the previous functional definitions. Given my degree of familiarity, I'm going to try this on the Lorenz system at the moment. If 3D equations prove to be a problem, we can find some other 1D equation. The Lorenz system has the following form:

\begin{equation}
\begin{aligned}
\dot{x} &= \sigma (y-x) \\
\dot{y} &= x (\rho - z) - y \\
\dot{z} &= xy - \beta z
\tag{3}
\end{aligned}
\end{equation}

```python
def lorenz_rhs(X, t, params):
    """
    Right hand side of the Lorenz system
    X : State vector
    t : Time
    params : Parameter list

    """
    x_dot = params[0] * (X[1] - X[0])
    y_dot = X[0]*(params[1] - X[2]) - X[1]
    z_dot = X[0]*X[1] - params[2]*X[2]
    return jnp.array([x_dot, y_dot, z_dot])

def solve_lorenz(X0, params, t_end, n_steps):
    """
    Solve the Lorenz system using the Runge-Kutta method
    X0 : Initial condition vector
    params : Parameter list (has form [sigma, rho, beta])
    t_end : End time
    n_steps : Number of time steps

    """
    ts = jnp.linspace(0, t_end, n_steps)
    Xs, ts_out = runge_kutta_solve(X0, lorenz_rhs, ts, params)

    return Xs, ts_out

solve_lorenz_batch_params = jax.vmap(solve_lorenz, in_axes=(None, 0, None, None), out_axes=(0,None))
solve_lorenz_batch_X0 = jax.vmap(solve_lorenz, in_axes=(0, None, None, None), out_axes=(0, None))

```

We can now simualte the lorenz equation for a variety of both initial conditions and a subset of parameters. We want to do parameter estimation using the neural net so the batch input into the neural net. So we need a function to create a set of random parameters [$\sigma$, $\rho$, $\beta$]. To do so we need two functions. One to generate the set of random parameters and one to use the set of random parameters to generate a set of random batch data using those random parameters. 

For the first function we need an input of a random key, the batchsize and a set of ranges

For the second we need use the parameter batch functions and the function notated above to generate a set of values. 

```python
def random_parameter_vector(key, batchsize, ranges):
    """
    Generate a batch of random parameters of size batchsize 
    within a given range for each parameter.
    key : JAX random key
    batchsize : Number of parameters to generate
    ranges : List of tuples defining (mix, max) for each parameter

    """
    params = []
    subkeys = jax.random.split( key, len(ranges) )
    for subkey, value_range in zip(subkeys, ranges):
        params.append(jax.random.uniform(subkey, [batchsize], minval = value_range[0], maxval = value_range[1]))
    return params

def lorenz_produce_training_batch(key, batchsize, ranges, t_end, n_steps, X0):
    """
    Produce a batch of training data with random parameters
    key : Random Jax key
    batchsize : Number of training examples to produce
    ranges : List of tuples defining (min, max) for each parameter
    t_end : End time for simulation
    n_steps : Number of time steps for simulation

    """
    sigmas, rhos, betas = random_parameter_vector(key, batchsize, ranges)
    params = jnp.stack([sigmas, rhos, betas], axis=1)
    Xs, ts = solve_lorenz_batch_params(X0, params, t_end, n_steps)
    return Xs, ts, params
```

```python
key = jax.random.key(42)
ranges = [(6.0,12.0), (20.,30.), (0.5,2.5)]  # sigma, rho, beta
batchsize = 100
tend = 40.0
n_steps = 4000
Xs, ts, params = lorenz_produce_training_batch(key, batchsize, ranges, 40.0, 4000, jnp.array([1.,1.,1.]))
print("Shape of Xs: ", Xs.shape) #(batchsize, n_steps-1, 3)
print("Shape of ts: ", ts.shape) # (n_steps-1,)
print("Shape of params: ", params.shape) # (batchsize, 3)
```

Now we define the Network is the same manner as Lessons 2 and 3. For now we don't need a convolutional network (I don't think) so we're going to define it normally as we did before, create the loss function, set the input and training parameters, initalize the network, and finally set up the training loop. 

The NN function takes in as an input a state and a list of parameters

The loss function takes in as an input a state, a target and a list of parameters and MSE losses the state and target

The loop runs over the total number of training batches, generates a new key, generates a new set of data, calculates the loss, updates the optimizer and the parameter list and appens the loss.

```python
def NN(x, params):
    """
    Simple feedforwards neural network

    x : Input vector
    params : List of weights and biases

    """
    num_layers = len(params['w'])
    for layer_idx, (w,b) in enumerate(zip(params['w'], params['b'])):
        x = jnp.matmul(w, x) + b
        if layer_idx < num_layers - 1:
           x = jnp.tanh(x)
    return x

NN_batch = jax.vmap(NN, in_axes=(0, None), out_axes=0)

def featurize_batch(Xs):
    # Xs: (B, T, 3)  ->  features: (B, T*3)
    mean = jnp.mean(Xs, axis=1, keepdims=True)
    std  = jnp.std(Xs,  axis=1, keepdims=True) + 1e-6
    Xn   = (Xs - mean) / std
    return Xn.reshape(Xn.shape[0], -1)

def mse_loss(nn_params, X_batch, y_target):
    """
    Mean squared error loss function
    x : Input vector
    y_target : Target vector
    params : List of weights and biases

    """
    feats = featurize_batch(X_batch)
    preds = NN_batch(feats, nn_params)
    loss = jnp.sum((preds - y_target)**2 ) 
    return loss

mse_loss_value_and_grad = jax.jit(jax.value_and_grad(mse_loss, argnums = 0))
```

```python
input_dim = 50
output_dim = 3
losses = []
#These are redefined here because I just wanted to test something above
key = jax.random.key(42)
ranges = [(6.0,12.0), (20.,30.), (0.5,2.5)]  # sigma, rho, beta
X0 = jnp.array([1.,1.,1.])
num_training_batches = 2000
batchsize = 100
tend = 40.0
n_steps = 4000
```

```python
optimizer = optax.adam(1e-3)

@jax.jit
def train_step(nn_params, opt_state, X_batch, y_target):
    loss, grads = mse_loss_value_and_grad(nn_params, X_batch, y_target)
    updates, opt_state = optimizer.update(grads, opt_state, nn_params)
    nn_params = optax.apply_updates(nn_params, updates)
    return nn_params, opt_state, loss
```

```python
key = jax.random.key(0)

# Shapes: D_in = (n_steps-1)*3 because your solver returns states at ts[1:]
D_in = (n_steps - 1) * 3
key, sub = jax.random.split(key)
nn_params = NN_init_params(sub, [D_in, 256, 128, 3])
opt_state = optimizer.init(nn_params)

losses = []

# one batch
key, sub = jax.random.split(key)
Xs, ts, y_target = lorenz_produce_training_batch(sub, batchsize, ranges, tend, n_steps, X0)
feats = featurize_batch(Xs)
assert feats.shape[1] == D_in
preds = NN_batch(feats, nn_params)          # (B,3)
print("preds shape:", preds.shape)          # -> (batchsize, 3)

# one train step
nn_params, opt_state, loss = train_step(nn_params, opt_state, Xs, y_target)
print("loss:", float(loss))
```
